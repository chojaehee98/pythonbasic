```python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')
titanic
```

```python
titanic.age.fillna(titanic.age.mean(), inplace =True)
bins = [1,20,30,50,70,100]
labels = ['미성년자','청년','장년','중년','노년']
```

```python
titanic[(titanic.age < 1) | (titanic.age > 100)]
```

```python
titanic_age = pd.cut(titanic.age,bins=bins,labels=labels).value_counts()
titanic_age
```

```python
titanic_age_rate = titanic_age/titanic_age.sum()
titanic_age_rate
```

```python
df_titanic_age = pd.DataFrame({'연령층': titanic_age, '연령 비율':titanic_age_rate}, index = labels)
df_titanic_age.loc['합계'] = df_titanic_age.sum(axis=0)
df_titanic_age.astype({'연령층':'int32'})
```

```python
# 진명님 풀이
bins = [1, 20, 30, 50, 70, 100]
labels = ["미성년자", "청년", "장년", "중년", "노년"]
titanic_age = pd.cut(titanic.age, bins, labels = labels)


titanic_age_df = pd.DataFrame(titanic_age.value_counts())
titanic_age_df.columns = ["연령층"]
titanic_age_df["연령비율"] = [round(i / titanic_age.count(), 2) for i in 
                      titanic_age.value_counts()]
titanic_age_df
```

```python
# 서현님 풀이
titanic.age = titanic.age.fillna(titanic.age.mean())
titanic_age = pd.cut(titanic.age, bins, labels = labels)
round(titanic_age.value_counts() / titanic_age.value_counts().sum(), 2)
```

```python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')
bins = [1, 20, 30, 50, 70, 100]
labels = ["미성년자", "청년", "장년", "중년", "노년"]
titanic.age.fillna(titanic.age.mean(), inplace =True)

titanic['연령층'] = pd.cut(titanic.age, bins, labels = labels)
titanic['category3'] = titanic.apply(lambda _: '미성년자' if _.age < 20 else _.연령층
                                     + ('남성' if _.sex=='male' else '여성'),axis=1)
titanic[['연령층','age','sex','category3']]
```
