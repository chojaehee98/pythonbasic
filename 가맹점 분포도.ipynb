```python
import json
import urllib.request

#한국 geojson html
kr_geojson = "https://raw.githubusercontent.com/southkorea/southkorea-maps/master/kostat/2018/json/skorea-municipalities-2018-geo.json"

#파싱
text_data = urllib.request.urlopen(kr_geojson).read().decode('utf-8')
geo_data = json.loads(text_data)
```

```python
#지역 이름 확인
[_['properties']['name'] for _ in geo_data['features']]
```

```python
import pandas as pd

#가맹점 데이터
gg_pay = pd.read_csv("datas/지역화폐가맹점(20230228기준).csv",encoding='CP949', low_memory = False)
```

```python
gg_pay = gg_pay[gg_pay["시군명"].notnull()]
gg_city = gg_pay["시군명"]
gg_city
```

```python
gg_city = pd.Series(gg_city.value_counts())
gg_city
```

```python
import folium

#지도
m = folium.Map(
    location = [37.57, 126.99],
    zoom_start = 11
)

#지도에 데이터 입히기
folium.Choropleth(
    geo_data=kr_geojson,
    data = gg_city,
    key_on = "features.properties.name",
    threshold_scale = [100, 300, 1000,3000, 10000, 40000],
    fill_color="YIGn",
    fill_opacity= 0.4,
    line_opacity=1,
    
).add_to(m)
m
```
