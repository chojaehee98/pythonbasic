```python
teams_result=pd.read_csv("datas/teams.csv").squeeze("columns")
teams_result["team"] = np.arange(28)
teams_result
```

```python
teams = pd.read_csv('datas/teams.csv')
o = list(range(1, 8))*4
np.random.shuffle(o)
teams['team'] = o
teams = teams.sort_values('team')
for i in range(teams.team.max()):
    temp = teams.loc[teams.team==(i+1)]['성명']
    print(f'{i+1}팀: ',list(temp))
```
