#### 남자 연령별 1인 연평균 지역화폐 사용 비용

```python
x = ['10대','20대','30대','40대','50대','60대','70대']
y = [a,b,c,d,e,f,g]

df = pd.DataFrame({'연령':x,'화폐':y}, columns=['연령','화폐'])
df
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FixedLocator, FixedFormatter

plt.title('남자 연령별 1인 연 평균 지역화폐 사용 비용')
ax = sns.barplot(x='연령',y='화폐',data=df,alpha=0.7)
# plt.yticks(rotation=90)
current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in current_values])

for patch in ax.patches:
    # 2개의 patch 값을 가짐
    ax.text(x = patch.get_x(),# + 100,
            y = patch.get_height(),
            s = f'{(patch.get_height()):1.0f}',
            ha = 'left')
plt.rc('font', size=11)
plt.show()
```
