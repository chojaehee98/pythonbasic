```python
import requests

from datetime import datetime
import time

def handle_rate_limits(response):
    now = datetime.now()
    reset_time = datetime.formtimestamp(int(response.headers['X-RateLimit_Reset']))
    remaining_requests = response.headers['X-Ratelimit-Remaining']
    remaining_time = (reset_time - now).total_seconds()
    intervals = remaining_time / (1.0 + int(remaining_requests))
    
    print('Sleepint for', intervals)
    time.sleep(intervals)
    time.sleep(intervals)
    return True
import xmltodict
import pandas as pd
for i in range(1,10):
    #time.sleep(2)
    response = requests.get('http://api.kcisa.kr/openapi/service/rest/meta/CHAheri',params={'serviceKey' : '451ac2f4-d15e-4925-b94d-1ae0af4c411c',"numOfRows":100,'pageNo':i})
    a = response.text
x = xmltodict.parse(str(a))['response']['body']['items']['item']    
p = pd.DataFrame(x)
p
```
