```python
Pokemons = pd.read_csv("datas/pokemon.csv", index_col="Pokemon").squeeze("columns")
Pokemons
```

```python
def single_or_multi(type):
    if '/' in type:
        return'Multi'
    return'Single'
```

```python
Pokemons.apply(single_or_multi)
```

```python
Pokemons.apply(single_or_multi).value_counts()
```

```python
import seaborn as sns
titanic=sns.load_dataset("titanic")
```

```python
titanic
```

```python
titanic['adult/child']= titanic.apply(lambda r: 'adult' if r.age >= 20 else 'child', axis=1)
titanic
```

```python
# (1)
titanic['category1'] = titanic.apply(lambda i: i['sex'] if i['age'] >= 20 else 'child',axis=1)
titanic.tail()
```

```python
# (2)
titanic['category1'] = titanic.apply(lambda r : 'child' if r.age<20 else r.sex, axis=1)
titanic.tail()
```
