All을 바꿔주기 위해 replace를 써서 변경
replace() 메서드의 첫 인수는 to_replace로 검색할 값을 지정하고
두번째 인수는 value로 변경할 값을 전달하면 된다

```python
inspections = inspections.replace(      # replace  
    to_replace = 'All', value = 'Risk 4 (Extreme)'
)
inspections['Risk'].unique()   # 통일성을 갖게 값 변경 완료
```

# DataFrame 합성
병합(merge)이나 연결을 지원

```python
df1 = pd.DataFrame({
    '고객번호': [1001,1002,1003,1004,1005,1006,1007],
    '이름': ['둘리','도우너','또치','길동','희동','마이콜','영희']
}, columns = ['고객번호','이름'])
df1
```

```python
df2 = pd.DataFrame({
    '고객번호': [1001,1001,1005,1006,1008,1001],
    '금액': [10000,20000,15000,5000,100000,30000]
}, columns = ['고객번호','금액'])
df2
```

양쪽 데이터프레임에 모두 키가 존재하는 데이터만 보여주는 inner join방식을 사용한다.  Merge()사용

```python
pd.merge(df1,df2)
```

how='outer' outer join: 키 값이 한쪽에만 있어도 데이터를 보여준다.

```python
pd.merge(df1, df2, how='outer')
```

left는 첫 번째 인수 기준으로 키 값을 모두 보여준다

```python
pd.merge(df1, df2, how='left')
```

right는 두번째 인수 기준으로 키 값을 모두 보여준다.

```python
pd.merge(df1, df2, how='right')
```

# DataFrame 그룹 분석
## 그룹 연산
### groupby()메서드

```python
np.random.seed(0)
df2 = pd.DataFrame({
    'key1':['A','A','B','B','A'],
    'key2':['one','two','one','two','one'],
    'data1':[1, 2, 3, 4, 5],
    'data2':[10, 20, 30, 40, 50]
})
df2
```

```python
groups = df2.groupby(df2.key1)
groups
```

```python
groups.groups
```

```python
groups.sum()
```

```python
df2.data1.groupby(df2.key1).sum()
```

```python
df2.groupby(df2.key1)['data1'].sum() # 위에꺼랑 같다
```

```python
df2.groupby(df2.key1).sum()['data1'] # 위에 두개랑 같다
```

```python
df2.data1.groupby([df2.key1, df2.key2]).sum()
```
