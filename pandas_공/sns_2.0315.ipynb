```python
def split_multicolumn(col_series):
    result_df = col_series.to_frame()
    options = []
    
    for idx, value in col_series[col_series.notnull()].items():
        for option in value.split(';'):     # split을 한 값을 for문을 통해 하나씩 보여줌
            if option not in result_df.columns:
                options.append(option)   # 값을 넣음 col_series안에
                result_df[option] = False # option 이라는 새로운 column을 생성해서 False 값을 넣어줌
            result_df.at[idx, option] = True # at[인덱스, column] 로 사용
    return result_df[options]
```

```python
dev_type_df = split_multicolumn(survey_df.DevType)
dev_type_df
```
